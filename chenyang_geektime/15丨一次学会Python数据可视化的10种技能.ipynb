## 可视化视图的种类

### 按数据间关系分类

比较：比较数据间各类别的关系，或者是它们随时间的变化趋势，比如<u>折线图</u>。

联系：查看变量之间的关系，比如<u>散点图</u>。

构成：每部分占整体的百分比，或随时间变化的百分比变化，比如<u>饼图</u>。

分布：关注单个变量，或多个变量的分布情况，比如<u>直方图</u>。

### 按变量个数分类

单变量：一个变量之间的关系。

多变量：两个或以上变量之间的关系。

（此处的「变量」可替换成「维度」。）

## 常见视图

### 散点图 Scatter Plot

特点：用来展示或探索变量之间的关系。

变种：气泡图。

常用场景：
- 关系数据探索
- 多变量映射，分析数据的相关性
- 与地图结合，生成地图散点图

#### Matplotlib 和 Seaborn 的绘制方法

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

N = 1000
x = np.random.randn(N)
y = np.random.randn(N)

# 使用 Matplotlib 绘图
plt.scatter(x, y, marker='x') # 不能使用英文半角双引号，只能使用英文半角单引号
plt.show()

# 使用 Seaborn 绘图
df = pd.DataFrame({'x': x, 'y': y})
sns.jointplot(x='x', y='y', data=df, kind='scatter')
plt.show()
```

### 折线图 Line Chart

特点：用来表示数据随时间变化的趋势。

变种：面积图。

常用场景：
- 数据随有序变量的变化趋势
- 分析多组数据随时间变化的相互作用和相互影响


#### Matplotlib 和 Seaborn 的绘制方法

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5, 3, 6, 20, 17, 16, 19, 30, 32, 35]

# 使用 Matplotlib 绘图
plt.plot(x, y)
plt.show()

# 使用 Seaborn 绘图
df = pd.DataFrame({'x':x, 'y':y})
sns.lineplot(x="x", y="y", data=df, color="red")
plt.show()
```

### 直方图 Histogram

特点：显示各组的频数或数量分布情况，从而显示差别。

变种：翻转的直方图。

常用场景：
- 表示分布情况
- 用于观察异常或孤立数据


#### Matplotlib 和 Seaborn 的绘制方法

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

a = np.random.randn(100)
s = pd.Series(a)

# 使用 Matplotlib 绘图
plt.hist(s)
plt.show()

# 使用 Seaborn 绘图
sns.distplot(s, kde=False)
plt.show()

sns.distplot(s, kde=True)
plt.show()
# kde 代表核密度估计，默认为 True
```

### 条形图（柱状图） Basic Bar Chart

特点：对比分类数据的大小。

变种：横向柱状图、堆叠柱状图、双轴柱状图。

常用场景：
- 分类数据对比（但分类数量不宜太多）


#### Matplotlib 和 Seaborn 的绘制方法

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = ['Cat1', 'Cat2', 'Cat3', 'Cat4', 'Cat5']
y = [5, 4, 8, 12, 7]

# 
plt.bar(x, y)
plt.show()

# 
sns.barplot(x, y)
plt.show()
```

### 箱线图（箱型图、盒须图） Box Chart

特点：显示数据的分布情况。

变种：

常用场景：
- 观察数据分布情况


#### Matplotlib 和 Seaborn 的绘制方法

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

data = np.random.normal(size=(10,4)) 
lables = ['A','B','C','D']

# 
plt.boxplot(data,labels=lables)
plt.show()

# 
df = pd.DataFrame(data, columns=lables)
sns.boxplot(data=df)
plt.show()
```

特点：显示数据的分布情况。

### 蜘蛛图（雷达图） Radar Chart（Spider Chart）

特点：显示各个变量的权重，以形成对比。

变种：

常用场景：
- 适合用于雷达图与雷达图之间的对比。


#### Matplotlib 和 Seaborn 的绘制方法

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties  

labels=np.array([u" 推进 ","KDA",u" 生存 ",u" 团战 ",u" 发育 ",u" 输出 "])
stats=[83, 61, 95, 67, 76, 88]

angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))

# 使用 Matplotlib 绘制
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)   
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)
font = FontProperties(fname=r"C:\Windows\Fonts\simhei.ttf", size=14)  
ax.set_thetagrids(angles * 180/np.pi, labels, FontProperties=font)
plt.show()
```

### 二元变量分布

特点：探索两个变量之间的关系。

变种：

常用场景：
- 数据探索。


#### Matplotlib 和 Seaborn 的绘制方法

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn 自带数据集 tips
tips = sns.load_dataset("tips")
print(tips.head(10))

# 使用 Seaborn 绘制
sns.jointplot(x="total_bill", y="tip", data=tips, kind='scatter')
sns.jointplot(x="total_bill", y="tip", data=tips, kind='kde')
sns.jointplot(x="total_bill", y="tip", data=tips, kind='hex')
plt.show()
```

### 成对关系

特点：探索数据集中的多个成对双变量的分布。

常用场景：
- 数据探索。
